# Exothermic Continuous Stirred Tank Reactor

Nonlinear dynamics of an exothermic continuous stirred-tank reactor
% (CSTR). This is example 2.05 from Seborg, et al.

## Model Equations

\begin{align*}
V\frac{dc_A}{dt} & = q(c_{Ai}-c_A)-Vkc_A \\
V\rho C_p\frac{dT}{dt} & = wC_p(T_i-T) + (-\Delta H_R)Vkc_A + UA(T_c-T)
\end{align*}

### Nominal Operating Conditions (Table 2.3)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

q = 100.0       # Flowrate [L/min]
cAi = 1.0       # Inlet feed concentration [mol/L]
Ti  = 350.0     # Inlet feed temperature [K]
V   = 100.0     # Volume [L]
rho = 1000.0    # Density [g/L]
Cp   = 0.239    # Heat capacity [J/g/K]
dHr = -5.0e4    # Heat of reaction [J/mol]

ER  = 8750.0    # Ea/R [K]
k0  = 7.2e10    # Arrhenius rate constant
UA  = 5.0e4     # Heat transfer [J/min/K]
Tc  = 300.0     # Coolant temperature [K]

cA0 = 0.5;      # Initial concentration [mol/L]
T0  = 350.0;    # Initial temperature [K]
```

## Simulation

```python
def k(T):
    return k0*np.exp(-ER/T)

def func(y,t):
    cA,T = y
    dcA = (q/V)*(cAi - cA) - k(T)*cA
    dT = (q/V)*(Ti - T) + (-dHr/rho/Cp)*k(T)*cA + (UA/V/rho/Cp)*(Tc-T)
    return [dcA,dT]
```

```python
IC = [0.5,350.0]
t = np.linspace(0,10.0,2000)
y = odeint(func,IC,t)

plt.figure()
plt.plot(t,y[:,0])
plt.xlabel('Time [min]')
plt.ylabel('Concentration [gmol/liter]')

plt.figure()
plt.plot(t,y[:,1])
plt.xlabel('Time [min]')
plt.ylabel('Temperature [K]')
```

## Effect of Cooling Temperature

```python
plt.figure(figsize=(12,4))
Tlist = [290,300,305]
for Tc in [290,300,305]:
    y = odeint(func,IC,t)

    plt.subplot(1,2,1)
    plt.plot(t,y[:,1])

    plt.subplot(1,2,2)
    plt.plot(t,y[:,0])
    
plt.subplot(1,2,1)
plt.legend(Tlist)
```

```python
from ipywidgets import interact

def sim(Tcooling):
    global Tc
    Tc = Tcooling
    y = odeint(func,IC,t)

    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(t,y[:,1])
    plt.xlabel('Time [min]')
    plt.ylabel('Temperature [K]')
    plt.ylim(300,460)
    plt.legend([Tc])
    
    plt.subplot(1,2,2)
    plt.plot(t,y[:,0])
    plt.xlabel('Time [min]')
    plt.ylabel('Concentration [gmol/liter]')
    plt.ylim(0,1)
    
interact(sim,Tcooling = (290.0,305.0))
```

## Nullclines

```python
# plot nullclines
T = np.linspace(300.0,460.0,1000)
    
plt.figure(figsize=(8,6))


plt.plot((q/V)*cAi/((q/V) + k(T)),T,'b')
plt.plot(((q/V)*(Ti-T) + (UA/V/rho/Cp)*(Tc-T))/((dHr/rho/Cp)*k(T)),T,'r')
plt.xlim(0,1)
plt.grid()
plt.legend(['dC/dt = 0','dT/dt = 0'])
plt.xlabel('Concentration')
plt.ylabel('Temperature')
```

## Phase Plane Analysis

```python
from ipywidgets import interact
from scipy.optimize import fsolve

def f(y,args):
    return func(y,0)

def phase(Tcooling):
    # put slider value into the global Tc
    global Tc
    Tc = Tcooling
    
    # plot nullclines
    T = linspace(300.0,460.0,1000)
    
    figure(figsize=(8,6))
    plot((q/V)*cAi/((q/V) + k(T)),T,'b--')
    plot(((q/V)*(Ti-T) + (UA/V/rho/Cp)*(Tc-T))/((dHr/rho/Cp)*k(T)),T,'r--')
    xlim(0,1)
    grid()
    
    # plot phase 
    y = odeint(func,IC,t)
    plot(y[:,0],y[:,1],'g',lw=2)
    
    legend(['dcA = 0','dT = 0','Phase'])
    xlabel('Concentration [gmol/liter]')
    ylabel('Temperature [K]')

interact(phase,Tcooling = (290.0,305.0))
```
