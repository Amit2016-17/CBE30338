# Hare and Lynx

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import interact

def LynxHare(H=20,L=20):
    a = 3.2
    b = 0.6
    c = 50
    d = 0.56
    k = 125
    r = 1.6

    def deriv(X,t):
        H,L = X
        Hdot =  r*H*(1-H/k) - a*H*L/(c+H)
        Ldot = b*a*H*L/(c+H) - d*L
        return [Hdot,Ldot]

    IC = [H,L]

    t = np.linspace(0,70,500)
    sol = odeint(deriv,IC,t)

    plt.figure(figsize=(12,6))
    plt.subplot(121)

    plt.plot(t,sol)

    plt.subplot(122)
    plt.plot(sol[:,0],sol[:,1])
    plt.xlim(0,130)
    plt.ylim(0,130)
    plt.title('Hare/Lynx Phase Plane')
    plt.xlabel('Hare')
    plt.ylabel('Lynx')
    
    Hp = np.linspace(0,k)
    Lp = r*(1-Hp/k)*(c+Hp)/a
    plt.plot(Hp,Lp)
    
    Hd = c*d/(a*b-d)
    plt.plot([Hd,Hd],plt.ylim())
```

```python
interact(LynxHare, H = (0,80,1), L =(0,80,1))
```
