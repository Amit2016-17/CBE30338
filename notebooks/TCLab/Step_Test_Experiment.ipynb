# Step Test: Experiment

## Executing the Step Test

### Verify an Initial Steady State

A step test assumes the system is initially at steady state. In the case of the Temperature Control Lab, the initial steady with no power input would be room temperature. We'll do a measurement to confirm the initial temperature.

```python
from tclab import TCLab, clock, Historian

a = TCLab()
print(a.T1,a.T1)
a.close()
```

### Conduct the Experiment

```python
# experimental parameters
Q1 = 50
tfinal = 500

# perform experiment
with TCLab() as a:
    h = Historian(a)
    h.initplot(tfinal)
    a.Q1(Q1)
    for t in clock(tfinal,strict=True):
        h.update()
```

### Verify the Experimental Data

```python
%matplotlib inline
import matplotlib.pyplot as plt

t = h.t
T1 = h.T1
T2 = h.T2
Q1 = h.Q1
Q2 = h.Q2

plt.plot(t,T1,t,T2,t,Q1,t,Q2)
plt.legend(['T1','T2','Q1','Q2'])
plt.xlabel('Time / seconds')
plt.grid()
```

### Convert to a DataFrame

```python
import pandas as pd

df = pd.DataFrame([t,T1,T2,Q1]).T
df.columns = ['Time','T1','T2','Q1']
df = df.set_index('Time')
df.plot(grid=True)
```

### Save Data as .csv file

```python
df.to_csv('StepTestData.csv')
```
