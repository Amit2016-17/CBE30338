# Dynamics of First Order Systems

```python
import pandas as pd

# Read data from spreadsheet data
df = pd.read_csv('ThermocoupleData.csv')
t = df['t']
T = df['T']
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# plot data
plt.figure(figsize=(12,5))
plt.scatter(t,T,s=20)
plt.title('Thermocouple Response')
plt.xlabel('Time [sec]')
plt.ylabel('Temperature [deg C]')
plt.xlim(min(t),max(t))
plt.grid()

plt.savefig('ThermocoupleResponse.png')
```

```python
plt.figure(figsize=(12,5))
plt.scatter(t,T,s=20)
plt.title('Thermocouple Response')
plt.xlabel('Time [sec]')
plt.ylabel('Temperature [deg C]')
plt.xlim(min(t),max(t))
plt.grid()

Tfit = 56 + (21-56)*np.exp(-t/0.45)
plt.plot(t,Tfit)

plt.savefig('ThermocoupleResponse.png')
```

```python
plt.scatter(t,T-Tfit)
plt.grid()
plt.title('Residuals')
```
